```python
import pandas as pd
import json
```

```python
# Load data from JSON file

with open('r_anxietyDepression.json', 'r') as f:
    data = json.load(f)
```

```python
data
```

```python
df_posts = pd.json_normalize(data)
df_posts
```

```python
# Normalize comments
df_comments = pd.json_normalize(data, record_path=['comments'], meta=['post_id', 'title', 'author', 'timestamp'], meta_prefix='comment_')

# Normalize replies within each comment
reply_data = []
for comment in data:
    if 'comments' in comment:
        for reply in comment['comments']:
            if 'replies' in reply:
                for reply_item in reply['replies']:
                    reply_item['comment_id'] = reply['comment_id']
                    reply_data.append(reply_item)

df_replies = pd.DataFrame(reply_data)
```
